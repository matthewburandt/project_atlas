```python
from fredapi import Fred
import pandas as pd
```

```python
from datetime import datetime
from datetime import timedelta
```

```python
fred_api_key = "d1a4e88cb84c34808b5c07e2e62b61ec"
```

```python
fred = Fred(api_key=fred_api_key)

# Define the series we want to collect
series = {
    "T10Y2Y": "Spread_2Y_10Y",
    "DFF": "Fed_Funds_Rate",
    "BAMLH0A0HYM2": "High_Yield_Spread",
    "VIXCLS": "VIX",
    "T10YIE": "Inflation_Expectations_10Y",
    "CPIAUCSL": "CPI",
    "UNRATE": "Unemployment_Rate"
}

# Set the date range for the data collection
period = timedelta(days=365) # last ~year of data
end_date = datetime.today()
start_date = end_date - period

history = pd.DataFrame()
for code, name in series.items():
    print(f"Collecting data for {name}...")
    data = fred.get_series(code, observation_start=start_date, observation_end=end_date)
    df = pd.DataFrame(data, columns=[name])
    df.index.name = "Date"
    history = pd.concat([history, df], axis=1)

history.index.name = "trade_date"
```
