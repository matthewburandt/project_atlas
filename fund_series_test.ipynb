```python
import yfinance as yf
import pandas as pd
```

```python
import datetime
from datetime import timedelta
```

```python

execution_date = datetime.datetime.today()
start = (execution_date - timedelta(days=1)).strftime('%Y-%m-%d')
end = execution_date.strftime('%Y-%m-%d')

etfs = ["SPY", "QQQ", "TLT", "GLD", "HYG"]

interval = "1d"

history = yf.download(etfs, start=start, end=end, interval=interval)

history.columns = ["_".join(col).strip() for col in history.columns.values]
history = history.rename(columns={'Date': 'trade_date'})
history = history.reset_index()
```
